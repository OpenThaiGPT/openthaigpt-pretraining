```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cls_df["tokenized_text"].tolist(), cls_df["labels"].tolist(), stratify = cls_df["labels"].tolist(),test_size=0.2, random_state=42)

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit(X_train)

train_features = vectorizer.transform(X_train).toarray()
test_features = vectorizer.transform(X_test).toarray()

from sklearn.linear_model import LogisticRegression

lr= LogisticRegression()

lr.fit(train_features,y_train)

prediction = lr.predict(test_features)

sum(prediction==y_test)/len(prediction)
```

# LDA

```python
f=open("/content/drive/MyDrive/Openthaigpt/stopwords.txt","r")

stopwords = [line.strip() for line in f.readlines()]

f.close()
```

```python
import gensim
from gensim import corpora
from gensim.models import LdaModel
from pythainlp.tokenize import word_tokenize
import re
import os
tqdm.pandas()

def preprocess_text(text):
  text = text.lower()
  text = text.replace("\n","")
  text = text.replace("\t","")
  text = re.sub("[^a-zA-Zก-๙๐-๙ ]","",text)
  tokenized_text = word_tokenize(text)
  return [w for w in tokenized_text if w not in stopwords and len(w)>1]
```

```python
tokenized_doc = df["text"].progress_apply(preprocess_text)


dictionary_path = "/content/drive/MyDrive/Openthaigpt/lda_dict_oscar2022"
check_file = os.path.isfile(dictionary_path)

if not check_file:
  dictionary = corpora.Dictionary(tokenized_doc)
  dictionary.save(dictionary_path)
else :
  dictionary=corpora.Dictionary.load("filename")


corpus = [dictionary.doc2bow(doc) for doc in tokenized_doc]

num_topics = 30
```

```python
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics , passes=20, iterations=100,random_state=1)
```

```python
for topic in lda_model.print_topics():
    print(topic)
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Visualize the topics
vis = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
docs_topic = []

for i in tqdm(range(len(df))):

  topics = lda_model.get_document_topics(corpus[i])

  top_topic = max(topics, key=lambda x: x[1])
  docs_topic.append(top_topic[0])
```
